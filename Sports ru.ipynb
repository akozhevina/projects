```python
empty_feature = 0
empty_obj = 0
for i in data.columns:
    if pd.Series.count(data[i]) < data.shape[0]:
        empty_feature += 1
for idx, row in data.iterrows():
    if pd.Series.count(row) < data.shape[1]:
        empty_obj += 1

print('количество объектов с пропусками: ', empty_obj)
print('количество признаков с пропусками: ', empty_feature)
```

Вычисление среднего, максимального/минимального значений, медианы 

```python
def features(dataset):
    df = pd.DataFrame(index = data.columns)
    df['min'] = dataset.min()
    df['max'] = dataset.max()
    df['mean'] = dataset.mean()
    df['median'] = dataset.median()
    return df
```

Для всего массива

```python
features(data)
```

```python
data.mode()
```

Для типа Photo

```python
df_ph = data[data['Type'] == 'Photo']
features(df_ph)
```

```python
df_ph.mode()
```

Для типа Status

```python
df_st = data[data['Type'] == 'Status']
features(df_st)
```

```python
df_st.mode()
```

Для типа Link

```python
df_link = data[data['Type'] == 'Link']
features(df_link)
```

```python
df_link.mode()
```

Для типа Video

```python
df_vid = data[data['Type'] == 'Video']
features(df_vid)
```
